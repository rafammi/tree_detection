```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.style.use("ggplot")
```

```python
results = pd.read_csv("/home/rafael/Projetos/tree_detection/results.csv")
```

```python
results = results[:-1]
```

```python
results
```

```python
melted = results.melt(id_vars="plot_id", value_vars=["recall",
"precision","f1"], var_name = "metric", value_name = "Value")
```

```python
melted
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='plot_id', y='Value', hue='metric', data=melted, palette='Set2')
plt.axhline(y = 0.571, color="#66c2a5", linestyle = "dashed", alpha = 0.8)
plt.axhline(y = 0.478, color="#fc8d62", linestyle = "dashed", alpha = 0.8)
plt.axhline(y = 0.501, color="#8da0cb", linestyle = "dashed", alpha = 0.8)
plt.ylabel('Score', fontsize=14)
plt.xlabel('Plot ID', fontsize=14)
plt.ylim(0, 1.0)
plt.savefig("results.png", dpi = 300)
```
