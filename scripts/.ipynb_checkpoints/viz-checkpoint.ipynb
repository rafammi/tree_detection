#### Field survey visualization (GT)

```python
gt = gpd.read_file("/home/rafael/Projetos/tree_detection/data/field_survey.geojson")
```

```python
gt.head(3)
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,8))
gt.plot(column="species", ax = ax, legend= True)
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,8))
gt.query("plot == 1").plot(column="species", ax = ax, legend= True)
```

```python
counts = gt.value_counts("species").reset_index()
counts

ax = sns.barplot(
    data=counts,
    y="count",
    x="species")
ax.set_title("Species distribution")
ax.grid(axis="x", color="black", alpha=0.1)
```

#### LAS files

```python
plot1 = laspy.read("/home/rafael/Projetos/tree_detection/data/als/plot_01.las").xyz

plot1 -= plot1.min(axis=0, keepdims=True)
```

```python
fig = plt.figure(figsize=(10, 5))

ax0 = fig.add_subplot(1, 1, 1, projection="3d")

p = ax0.scatter(
    *plot1.swapaxes(0, 1),
    c=plot1[:, 2],  # color by height (Z)
    cmap="viridis", # pick a colormap with contrast
    s=2
)
ax0.view_init(elev=30, azim=45)

ax0.set_title("Plot 1", y=0.83)
ax0.set_aspect("equal")
fig.colorbar(p, ax=ax0, label="Height")
```
